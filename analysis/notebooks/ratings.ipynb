```python
import pandas as pd
import os
import xlsxwriter
```

```python
responses_csv = "results-survey578575-120completed-code-wo-questiontimes.csv"
responses_df = pd.read_csv(os.path.join("./input", responses_csv))
```

```python
print(responses_df)
```

```python
header_row_codetext = list(responses_df)
header_row_code = [header.split(". ")[0] for header in header_row_codetext]

header_dict = {}
for header in header_row_codetext:
    splitted_header = header.split(". ")
    print(splitted_header)
    if len(splitted_header) > 1:
        header_dict[splitted_header[0]] = splitted_header[1]

print(header_dict)
```

```python
print(header_row_code)
```

```python
new_header = dict(zip(header_row_codetext, header_row_code))
#print(new_header)
responses_df.rename(columns=new_header, inplace=True)
```

```python
print(responses_df)
```

```python
usefulness = [
    "G09Q24[SQ001]", "G09Q26[SQ001]", "G09Q28[SQ001]",
    "G09Q30[SQ001]", "G09Q32[SQ001]", "G09Q34[SQ001]", "G09Q36[SQ001]", "G09Q38[SQ001]",
    "G09Q40[SQ001]", "G09Q42[SQ001]", "G09Q44[SQ001]", "G09Q46[SQ001]", "G09Q48[SQ001]",
    "G09Q50[SQ001]", "G09Q52[SQ001]", "G09Q54[SQ001]", "G09Q56[SQ001]", "G09Q56[SQ001]"
]

questions = []
stats_df = pd.DataFrame()

writer = pd.ExcelWriter("./output/ratings.xlsx", engine="xlsxwriter")
for sheet in usefulness:
    question = header_dict[sheet]

    tmp_df = responses_df[[sheet, "PROLIFICPID"]]
    responses = list(tmp_df[sheet])
    prolific_ids = list(tmp_df["PROLIFICPID"])

    # split "5 (very useful)" by spaces, take the first element as the rating, here: 5
    responses_values = [int(value.split(" ")[0]) for value in responses]
    s = pd.Series(responses_values)
    statistical_desc = s.describe()
    statistical_desc = pd.DataFrame(statistical_desc)
    statistical_desc = statistical_desc.transpose()

    questions.append(question)
    stats_df = pd.concat([stats_df, statistical_desc])

    df = pd.DataFrame()
    df["PROLIFICPID"] = prolific_ids
    df[question] = responses_values

    header_df = list(df)
    header_stats = list(statistical_desc)
    df = pd.concat([df, statistical_desc], axis=1, ignore_index=True)
    df.columns = header_df + header_stats

    print(df)
    sheet_name = sheet.split("[")[0]
    df.to_excel(writer, sheet_name=sheet_name)

writer.save()

stats_df.insert(loc=0, column='Questions', value=questions)
writer = pd.ExcelWriter("./output/rating_stats.xlsx", engine="xlsxwriter")
stats_df.to_excel(writer, sheet_name="Rating Statistics Overview")
writer.save()

```

```python
frequency = [
    "G02Q05[SQ001]", "G02Q04[SQ001]",
    "G02Q07[SQ01]", "G02Q07[SQ02]",
    "G02Q08[SQ01]", "G02Q08[SQ02]", "G02Q08[SQ03]", "G02Q08[SQ04]",
    "G02Q09[SQ01]", "G02Q09[SQ02]", "G02Q09[SQ03]", "G02Q09[SQ04]",
    "G01Q11[SQ01]",
    "G02Q12[SQ01]", "G02Q12[SQ02]", "G02Q12[SQ03]", "G02Q12[SQ04]", "G02Q12[SQ05]",
    "G03Q14[SQ01]", "G03Q14[SQ02]", "G03Q14[SQ03]", "G03Q14[SQ04]", "G03Q14[SQ05]", "G03Q14[SQ06]", "G03Q14[SQ07]",
    "G03Q16[SQ01]", "G03Q16[SQ02]", "G03Q16[SQ03]", "G03Q16[SQ04]", "G03Q16[SQ05]", "G03Q16[SQ06]", "G03Q16[SQ07]"
]

questions = []
stats_df = pd.DataFrame()

writer = pd.ExcelWriter("./output/frequency.xlsx", engine="xlsxwriter")
for sheet in frequency:
    question = header_dict[sheet]

    tmp_df = responses_df[[sheet, "PROLIFICPID"]]
    responses = list(tmp_df[sheet])
    prolific_ids = list(tmp_df["PROLIFICPID"])

    responses_values = []
    for r in responses:
        if (r == "daily"):
            responses_values.append(5)
        elif (r == "weekly"):
            responses_values.append(4)
        elif (r == "monthly"):
            responses_values.append(3)
        elif (r == "less than monthly"):
            responses_values.append(2)
        elif (r == "never"):
            responses_values.append(1)
    # split "5 (very useful)" by spaces, take the first element as the rating, here: 5
    s = pd.Series(responses_values)
    statistical_desc = s.describe()
    statistical_desc = pd.DataFrame(statistical_desc)
    statistical_desc = statistical_desc.transpose()

    questions.append(question)
    stats_df = pd.concat([stats_df, statistical_desc])
    #print(responses_values)

    #print(tmp_df)

    df = pd.DataFrame()
    df["PROLIFICPID"] = prolific_ids
    df[question] = responses_values

    header_df = list(df)
    header_stats = list(statistical_desc)
    df = pd.concat([df, statistical_desc], axis=1, ignore_index=True)
    df.columns = header_df + header_stats

    print(df)
    sheet_name = sheet.replace("[", "_")
    sheet_name = sheet_name.replace("]", "")
    df.to_excel(writer, sheet_name=sheet_name)

writer.save()

stats_df.insert(loc=0, column='Questions', value=questions)
writer = pd.ExcelWriter("./output/frequency_stats.xlsx", engine="xlsxwriter")
stats_df.to_excel(writer, sheet_name="Frequency Statistics Overview")
writer.save()
```
